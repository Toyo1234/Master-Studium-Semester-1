## Aufgabe 2

**Downloaden und Laden Sie die Datei Nashorn.jpg oder ein eigenes Bild**

Ergänzen Sie den folgenden Python Code, so dass ein größeres OpenCV Grauwertbild
(Default: Nashorn.jpg) invertiert wird ```(neues_Matrixelement = 255 – altes_Matrixelement)```. Das Bild
ist eigentlich ein numpy.array. Versuchen Sie unterschiedliche Arten des Zugriffs, benutzen Sie:

````python
import time
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

def plot(img):
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    plt.show()

imgPath = 'Nashorn.jpg'
img = cv.imread(imgPath , cv.IMREAD_GRAYSCALE)
assert img is not None, imgPath + " nicht gefunden"
(rows, cols) = img.shape
nPix = rows * cols
img2 = np.zeros(img.shape, dtype=np.uint8)
plot(img)
````

### **a)**
Mittels 2 geschachtelter Schleifen (über Zeilen und Spalten) und:

```neu = 255 - img[row, col]```

````python
# Geschachtelte Schleifen über Zeilen und Spalten
# img[r, c] = 255 - img[r, c]
nLoops = 1
start = time.perf_counter()
for i in range(nLoops):
    img2 = np.zeros(img.shape, dtype=np.uint8)
    for row in range(rows):
        for col in range(cols):
            img2[row, col] = 255 -img[row,col]
dt = (time.perf_counter() - start) / nLoops
print("Geschachtelte Schleifen: {:5.2f}s, {:5.3f}Mpix/s"
      .format(dt, 1e-6*nPix/dt))
plot(img2)
````

````python
### **b)**
#Mittels Matrixoperation:
nLoops = 10
start = time.perf_counter()
for i in range(nLoops):
    img2 = 255 - img
dt = (time.perf_counter() - start) / nLoops
print("Matrixoperation: {:5.3f}ms, {:5d}Mpix/s"
      .format(dt*1000, int(1e-6*nPix/dt)))
plot(img2)
````

### **c)**
Durch Definieren einer Lookup-Table
```(np.array)``` mit Zahlen ```255, 254, …, 0```.
Anwenden der Lookup Table mit:

```cv.LUT(...)```

````python
# Lookup Table Operation (optimierter Code in OpenCV)
lut = 255 - np.arange(256, dtype=np.uint8)
nLoops = 10
start = time.perf_counter()
for i in range(nLoops):
    # TODO: ===== Hier eigener Code =====
    img2 = cv.LUT(img, lut)

dt = (time.perf_counter() - start) / nLoops
print("Lookup Table Operation:  {:5.3f}ms, {:5d}Mpix/s"
      .format(dt*1000, int(1e-6*nPix/dt)))
plot(img2)
````
