### Beispiel

````python
# Import von Bibliotheken
import numpy as np
import matplotlib.pyplot as plt

# Berechnung
np.random.seed(0)
n = 100000
x = np.random.standard_normal(n)
y = 2.0 + 3.0 * x + 4.0 * np.random.standard_normal(n)
xmin = x.min()
xmax = x.max()
ymin = y.min()
ymax = y.max()

# Ausgabe
fig, ax = plt.subplots(figsize=(7,4))
hb = ax.hexbin(x, y, gridsize=50, bins='log', cmap='inferno')
ax.axis([xmin, xmax, ymin, ymax])
ax.set_title("log color scale")
cb = fig.colorbar(hb, ax=ax)
cb.set_label('log10(N)')
plt.show()
````

Interaktive 3D-Visualisierungen — das Diagramm lässt sich mit der Maus drehen und zoomen.
Dafür wird das `ipympl`-Backend aktiviert (`conda install ipympl`).
Ohne es fällt Jupyter auf eine statische Ausgabe zurück.

````python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Gitter erstellen
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# 2D-Gaußglocke — typische Form in ML (z. B. RBF-Kernel, Wahrscheinlichkeitsdichten)
Z = np.exp(-(X**2 + Y**2) / 2)

fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('2D-Gaußglocke')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.tight_layout()
plt.show()
````

#### `None` und `NoneType`

`None` ist Pythons Pendant zu `null` / `void` — es repräsentiert das **Fehlen eines Wertes**.
Funktionen ohne explizites `return` geben implizit `None` zurück.

````python
x = None
print(x)
print(type(x))
````

#### `==` vs. `is`

| Operator | Prüft ... |
|:---------|:----------|
| `==`     | **Wertgleichheit** — sind die Inhalte gleich? |
| `is`     | **Identität** — zeigen beide Namen auf dasselbe Objekt im Speicher? |

Faustregel: `is` nur für `None`, `True` und `False` verwenden.

````python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)  # True  — gleicher Inhalt
print(a is b)  # False — verschiedene Objekte im Speicher
````

````python
c = a           # c zeigt auf dasselbe Objekt wie a
print(a is c)  # True
````

````python
# Standardfall: None-Prüfung immer mit 'is'
result = None
if result is None:
    print("Kein Ergebnis vorhanden")
````

### Lambda-Funktionen

Eine Lambda-Funktion ist eine **anonyme Einzeiler-Funktion** — nützlich überall,
wo eine kleine Funktion nur einmal gebraucht wird.

```
lambda <parameter> : <ausdruck>
```

````python
square = lambda x: x ** 2
square(5)
````

````python
# Äquivalente normale Funktion — bei mehrfacher Nutzung bevorzugen
def square(x):
    return x ** 2
````

````python
# Häufiger Einsatz: Sortierschlüssel
punkte = [("Alice", 92), ("Bob", 85), ("Carol", 97)]
sorted(punkte, key=lambda p: p[1])
````

````python
# map() — Funktion auf jedes Element anwenden
werte = [1, 2, 3, 4, 5]
list(map(lambda x: x ** 2, werte))
````

````python
# filter() — Elemente behalten, für die die Funktion True zurückgibt
list(filter(lambda x: x % 2 == 0, werte))
````

### Klassen

````python
class Point:
    zero = 0 # Klassenvariable
    
    def __init__(self, x, y):
        # Objektvariablen
        self.x = x
        self.y = y
        
class Line:
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
        
    def length(self):
        a = self.p1
        b = self.p2
        return ((a.x - b.x)**2 + (a.y - b.y)**2)**0.5
````

````python
p1 = Point(1, 1)
p2 = Point(Point.zero, 2)
````

````python
line = Line(p1, p2)
````

````python
line.length()
````

### Module und Imports

Python-Code ist in **Module** (einzelne `.py`-Dateien) und **Pakete** (Verzeichnisse mit Modulen) organisiert.
Mit `import` wird ein Modul geladen und sein Inhalt zugänglich gemacht.

| Syntax | Bedeutung |
|:-------|:----------|
| `import modul` | Modul laden, Zugriff über `modul.name` |
| `import modul as alias` | Modul unter kurzem Namen einbinden |
| `from modul import name` | Einzelnes Objekt direkt importieren |
| `from modul import *` | Alles importieren — **nicht empfohlen** |

````python
import math
math.sqrt(16)
````
